## Some Data Cleaning

Let's clean us some data.

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

The raw data. We have both the full Books of Nonesense and also just the limericks:

```python
bon = []
with open("../data/limericks.txt", 'r') as f:
    for line in f:
        if not "[Illustration]" in line:
            bon.append(line.rstrip().lstrip())
```

```python
bon
```

```python
a = "This has a Don\'t character"
b = a.strip("\\")
```

```python
limericks_all = []
for i,b in enumerate(bon[:-4]):
    if len(b) == 0:
        if len(bon[i+1]) == 0:
            continue
        else:
            b1 = bon[i+1].strip("\\") + "\n"
            b2 = bon[i+2].strip("\\") + "\n"

            b3 = bon[i+3].strip("\\") + "\n"
            b4 = bon[i+4].strip("\\") + "\n"
            
            limerick = " ".join([b1, b2, b3, b4])
            limericks_all.append(limerick)
    else:
        continue
```

```python
limericks_all
```

```python
limericks_all = pd.Series(limericks_all)
```

```python
limericks_all.to_json("../data/limericks_all.json")
```
