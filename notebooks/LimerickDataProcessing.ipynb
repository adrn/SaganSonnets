## Some Data Cleaning

Let's clean us some data.

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

The raw data. We have both the full Books of Nonesense and also just the limericks:

```python
bon = []
with open("../data/booksofnonsense.txt", 'r') as f:
    for line in f:
        if not "[Illustration]" in line:
            bon.append(line.rstrip().lstrip())
```

```python
bon
```

```python
with open("../data/booksofnonsense_cleaned.txt", "w") as fnew:
    for line in bon:
        fnew.write(line + "\n")
```
