```python
import bs4
import tqdm
import requests
```

```python
templ = 'http://www.feynmanlectures.caltech.edu/{volnum}_{chnum:02}.html'
num_to_roman = {1:'I', 2:'II', 3:'III'}
def make_url(volnum, chnum):
    return templ.format(volnum=num_to_roman[volnum], chnum=chnum)

nchapters = {1:52, 2:42, 3: 21}
```

```python
def extract_paragraphs(bs4doc):
    chapters = bs4doc.findAll(attrs={'class':'chapter'})
    assert len(chapters) == 1
    chapter = chapters[0]
    
    paras = [t.get_text().strip().replace('\r', '') for t in chapter.findAll(attrs={'class': 'para'})]
    return '\n\n'.join(paras)
```

# Extract all text 

```python
volchs = []
for volnum in range(1,4):
    for i in range(nchapters[volnum]):
        volchs.append((volnum, i+1))

texts = {}
erred = {}
for volch in tqdm.tqdm_notebook(volchs):
    volnum, chnum = volch
    req = requests.get(make_url(*volch))
    if not req.ok:
        print('Volch', volch, 'failed with code', req.status_code)
        erred[(volnum, chnum)] = req
    else:
        bs4doc = bs4.BeautifulSoup(req.text)
        texts[volch] = extract_paragraphs(bs4doc)
```

```python
with open('../data/feynman_lectures_paras.txt', 'w') as f:
    for text in texts.values():
        f.write(text)
        f.write('\n\n')
```
